```python
contents = [line.rstrip('\n') for line in open('content.txt')]
print(contents)
```

```python
contents_tag = []
for c in contents:
    contents_tag.append(c.split(' '))
    
distinct_words = set(word for document in contents_tag for word in document)
W = len(distinct_words)
ss = sorted(distinct_words)
for _ in ss:
    print(_)
```

```python
documents = []
for tweet in contents:
    tweet_split = tweet.split(' ')
    
    temp = []
    for word in tweet_split:
        w = word.split("/", 1)[0]
        temp.append(w)
    documents.append(temp)
```

```python
distinct_words_doc = set(word for document in documents for word in document)
W1 = len(distinct_words_doc)
s = sorted(distinct_words_doc)
for _ in s:
    print(_)
```

```python
import nltk.tag

doc_tup = []
for tweet in contents:
    word = tweet.split(' ')
    
    tup = []
    for w in word:
        tup.append(nltk.tag.str2tuple(w))
    doc_tup += [tup]

# print(doc_tup)
```
