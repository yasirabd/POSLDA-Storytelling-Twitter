```python
import pandas as pd

df = pd.read_csv('kidney_disease.csv')
df.head()
```

```python
numeric_df = df.drop(['id','rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'classification'], axis=1)
numeric_df.head()
```

```python
numeric_df['pcv'] = pd.to_numeric(numeric_df['pcv'], errors='coerce')
numeric_df['wc'] = pd.to_numeric(numeric_df['wc'], errors='coerce')
numeric_df['rc'] = pd.to_numeric(numeric_df['rc'], errors='coerce')
```

```python
lower = 0.05
upper = 0.95

# print(type(numeric_df['pcv'][0]))
quant_df = numeric_df.quantile([lower, upper])
quant_df
```

```python
# conver low outliers
numeric_df = numeric_df.mask((numeric_df < df.quantile(0.05)), numeric_df.quantile(0.05), axis=1)  

# conver low outliers
numeric_df = numeric_df.mask((numeric_df > df.quantile(0.95)), numeric_df.quantile(0.95), axis=1)  
numeric_df.head()
```
