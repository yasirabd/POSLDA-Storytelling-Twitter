```python
import os
import time
import pandas as pd
import numpy as np
from collections import defaultdict
from hmmtagger.tagger import MainTagger
from tokenization import *

mt = None
```

```python
Ccon = ['JJ', 'NN','NNP', 'NNG', 'VBI', 'VBT', 'FW']
Cfunc = ['OP', 'CP', 'GM', ';', ':', '"', '.', 
         ',', '-', '...', 'RB', 'IN', 'MD', 'CC',
         'SC', 'DT', 'UH', 'CDO', 'CDC', 'CDP', 'CDI',
         'PRP', 'WP', 'PRN', 'PRL', 'NEG', 'SYM', 'RP']
```

```python
Kelas = {"Ccon": {},
         "Cfunc": defaultdict(list)
        }
```

```python
df_normalize = pd.read_csv('normalize_export.csv', header=None, names=['normalize'])
```

```python
df_normalize.head()
```

```python
docs = df_normalize.values.tolist()
docs
```

```python
"""
Parameters:
    String fileLexicon
    String fileNGram
    int NGramType = 0
    int maxAffixLength = 3
    int Treshold = 3
    int minWordFreq = 0
    int modeAffixTree = 0
    boolean debug = False
    double LambdaBigram = 0.2
    int TwoPhaseType = 0
    double beamFactor = 500.0
    int useLexicon = 0
"""

def init_tag():
    global mt
    try:
        if mt is None:
            mt = MainTagger("resource/Lexicon.trn", "resource/Ngram.trn", 0, 3, 3, 0, 0, False, 0.2, 0, 500.0, 1)
    except:
        print("Error Exception")
```

```python
%%time

idx = 0
# tokenisasi dan tagging
tagged_docs = []
for doc in docs:
    start = time.clock()
    try:
        init_tag()
        for d in doc:
            if len(d) == 0: continue
            out = sentence_extraction(cleaning(d))
            join_token = []
            for o in out:
                strtag = " ".join(tokenisasi_kalimat(o)).strip()
                join_token.extend(mt.taggingStr(strtag))
            tagged_docs += [join_token]
        print("doc", idx, "selesai", time.clock()-start, "seconds")
        idx += 1
    except ValueError:
        print ("Error Exceptions")
```

```python
for _ in tagged_docs:
    print (_)
```

```python
import nltk.tag
```

```python
doc_con = []
for tagged in tagged_docs:
    con = []
    for word in tagged:
        w = word.split("/", 1)[0]
        tagger = word.split("/", 1)[1]
        if tagger in Ccon:
            con += ["".join(word)]
        elif tagger in Cfunc:
            Kelas["Cfunc"][tagger].append(w)
    doc_con += [con]
```

```python
for key in Kelas["Cfunc"]:
    Kelas["Cfunc"][key] = list(set(Kelas["Cfunc"][key]))
print(Kelas)
```

```python
documents = []
for con in doc_con:
    co = []
    for c in con:
        result = c.split('/', 1)[0]
        co.append(result)
    documents += [co]
```

```python
for _ in documents:
    print (_)
```

```python
doc_tup = []
for _ in doc_con:
    tup = []
    for w in _:
        tup.append(nltk.tag.str2tuple(w))
    doc_tup += [tup]

for _ in doc_tup:
    print(_)
```
